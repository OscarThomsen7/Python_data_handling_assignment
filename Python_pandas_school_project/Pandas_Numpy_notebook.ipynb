<font size = 7>PART 3. VG: Analys</font>

```python
data = pd.read_csv('FT1000.csv')
url = 'https://www.kaggle.com/datasets/nguyenthicamlai/top-1000-fastest-growing-companies-in-europe'
data = data.dropna().drop_duplicates()
```

<font size = 5>1. Vilka sectors hade högst genomsnittlig revenue 2020?</font>

```python
median_revenue = pd.DataFrame(data.groupby(['Sector']).mean('Revenue2020'))
median_revenue = median_revenue['Revenue2020'].sort_values(ascending=False)
plt.rcParams['figure.figsize'] = [15, 6]
plt.title('Bar chart that shows the median revenue of each sector in 2020')
plt.ylabel('Revenue from 2020 in Euros (100 Million)')
median_revenue.plot.bar()
plt.show()
```

<font size = 5>2. Vilka är de 10 företagen med högst CAGR från 2017 till 2020?</font>

```python
top_10 = pd.DataFrame(data.nlargest(10, 'CAGR'))
highest_revenue = pd.DataFrame(top_10.groupby(['Name']).sum('CAGR'))
highest_revenue = highest_revenue['CAGR'].sort_values(ascending=False)
plt.title('Bar chart that shows the top 10 companies CAGR from 2017 to 2020')
plt.ylabel('CAGR from 2017 to 2020')
plt.gca().yaxis.set_major_formatter(mtick.PercentFormatter(xmax=100.0))
highest_revenue.plot.bar()
plt.show()
```

<font size = 5>3. Jämför hur stor andel av företagen som var med i listan 2021 och hur stor andel som inte var med.</font>

```python
not_ranked = data.loc[data['Ranked2021'] == 'No'].count()[0]
ranked = data.loc[data['Ranked2021'] == 'Yes'].count()[0]
plt.title('Pie chart of all the companies ranked status in 2021')
labels = ['Not Ranked', 'Ranked']
plt.pie([not_ranked, ranked], labels = labels, autopct='%1.2f%%')
plt.show()
```

<font size = 5>4. Jämför hur många anställda de olika sektorerna hade 2020 och 2017.</font>

```python
total_employees = data.groupby(['Sector'])[['Employees2020', 'Employees2017']].sum()
plt.xticks(rotation = 90)
plt.xlabel('Sector')
plt.ylabel('Employees')
plt.title('Comparison between sectors employees from 2017 and 2020')
labels = ['Total employees 2020', 'Total employees 2017']
plt.plot(total_employees)
plt.legend([labels[0], labels[1]])
plt.show()
```
